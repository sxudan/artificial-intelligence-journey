```python
import librosa
```

```python
input_audio, sr = librosa.load('audio.wav', sr=22050)

```

```python
librosa.get_samplerate(path='audio.wav')
```

```python
import IPython
import IPython.display

IPython.display.Audio(input_audio,rate=22050)
```

```python
spectrogram = librosa.stft(input_audio)
```

```python
print(spectrogram.shape)
```

```python
spectrogram_db = librosa.amplitude_to_db(abs(spectrogram))
```

```python
import matplotlib.pyplot as plt
print(spectrogram_db)
plt.figure(figsize=(10, 4))
# librosa.display.specshow(spectrogram_db, sr=22050, x_axis='time', y_axis='linear')
plt.imshow(spectrogram_db, origin="lower", aspect="auto")
plt.colorbar(format='%+2.0f dB')
plt.title('Spectrogram')
plt.xlabel('Time (s)')
plt.ylabel('Frequency (Hz)')
plt.tight_layout()
plt.show()
```

```python
import torchaudio
import torch

device = torch.device('cpu')

waveform = librosa.griffinlim(spectrogram)
```

```python
print(waveform.shape)
plt.plot(waveform)
```
