```python
import torchaudio
import IPython.display
```

```python
waveforms, sr = torchaudio.load('audio.wav')
```

```python
import matplotlib.pyplot as plt

plt.plot(waveforms[0])

IPython.display.Audio(waveforms, rate=sr)
```

```python
spec = torchaudio.transforms.Spectrogram(1024, power=2)


apl2db = torchaudio.transforms.AmplitudeToDB()

spectogram = spec(waveforms)
spectogram_db = apl2db(spectogram)

print(spectogram_db.shape)

plt.imshow(spectogram_db[0], origin="lower", aspect="auto")
```

```python
inverse = torchaudio.transforms.GriffinLim(1024, power=2)

wv = inverse(spectogram)

IPython.display.Audio(wv, rate=sr)
```
